```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
filter1 = df[['Name', 'Age', 'Sex', 'Pclass', 'Survived']]
filter1 = filter1.loc[(df['Sex'] == 'female') & (df['Pclass'] == 1) & (df['Survived'] == 1)].sort_values('Name')
```

```python
filter1.to_csv('comparacao.csv', index=False)
```
